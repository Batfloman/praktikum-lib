```python
import numpy as np

x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.2, 4.9, 7.8, 10.2, 12.9])
yerr = np.array([0.3, 0.4, 0.3, 0.4, 0.3])
xerr = np.array([0.1, 0.2, 0.1, 0.2, 0.1])
```

```python
a = graph_fit.Linear.fit(x_data, y_data)
print(a.params)
```

```python
b = graph_fit.Linear.fit(x_data, y_data, yerr=yerr)
print(b.params)
```

```python
c = graph_fit.Linear.odr_fit(x_data, y_data, xerr, yerr)
print(c.params)
```

```python
plot = graph.create_plot()
fig, ax = plot

graph.scatter(x_data, y_data, plot=plot)

for i, fit in enumerate([a,b,c]):
    graph.plot_func(fit, plot=plot, change_viewport=False, label=f"fit {i}")

ax.legend()
```
